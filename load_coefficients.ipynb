```python
def number_of_elements(number_of_pixels: int):
    return ceil((2*ceil(number_of_pixels/(2**(0.5)))+1)/2)

def load_variables(img_size: int, noise_level: float, model: str):
    dir = f"./LearnedRadonFilters/det3_22_11_12/epochs=5, dataset.img_size={img_size}, noise_level={noise_level}, model={model}"
    loaded_pi = torch.load(f"{dir}\pi.pt")
    loaded_gamma = torch.load(f"{dir}\gamma.pt")
    loaded_delta = torch.load(f"{dir}\delta.pt")


    loaded_pi = torch.nn.utils.parameters_to_vector(loaded_pi).view(loaded_pi.shape).cpu()
    loaded_gamma = torch.nn.utils.parameters_to_vector(loaded_gamma).view(loaded_gamma.shape).cpu()
    loaded_delta = torch.nn.utils.parameters_to_vector(loaded_delta).view(loaded_delta.shape).cpu()
    loaded_count = torch.load(f"{dir}\count.pt")

    if "learned" in model:
        loaded_pi /= 5
        loaded_gamma /= 5
        loaded_delta /= 5
        loaded_count /= 5
    

    loaded_g = torch.load(f"{dir}\coefficients.pt").detach().cpu()
    return loaded_pi, loaded_gamma, loaded_delta, loaded_count, loaded_g


```
